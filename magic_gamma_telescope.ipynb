```python
# Magic Gamma Telescope

# ---------- Used libraries:
import numpy as np
# ---------- used for handling numbers
import pandas as pd
# ---------- used for handling the dataset
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
# ---------- used for normalizing data
from sklearn.preprocessing import StandardScaler
# ---------- used for cross validation
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score, cross_val_predict
# ---------- used for linear model
from sklearn import linear_model
# ---------- used for bagging
from numpy import mean
from numpy import std
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import BaggingClassifier
# ---------- used for boosting
from sklearn.ensemble import AdaBoostClassifier
# ---------- used for decision tree
from sklearn.tree import DecisionTreeClassifier
# ---------- used for naive bayes
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
# ---------- used for artificial neural network
from sklearn.neural_network import MLPClassifier
# ---------- used for svm
from sklearn import svm

####
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
###
import time
import matplotlib.pyplot as plt 
from sklearn.metrics import plot_confusion_matrix
import seaborn as sn


# ---------------- Data PreProcessing ----------------------------

dataset = pd.read_csv('magic.csv')  # import the dataset into a variable

dataset.info()  # information about the dataset
dataset = shuffle(dataset)  # shuffling the data

# Replacing 'g' values with 1 and 'h' values with 0
dataset['class'] = dataset['class'].replace(['g'], 1)
dataset['class'] = dataset['class'].replace(['h'], 0)

# Splitting the attributes into independent and dependent attributes
X = dataset[['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist']] # features
Y = dataset['class']   # label

#Decide whether the dataset is balanced or imbalanced by comparing the number of 1's and 0's in the target data
count_1 = 0 
count_0 = 0
for i in Y:
    #target 'h'
    if (i == 0):
        count_0 += 1
    #target 'g'
    else:
        count_1 += 1

print("Number of samples with target value 'g':", count_1)
print("Number of samples with target value 'h':", count_0)
        
# Normalizing the values
X = StandardScaler().fit_transform(X)

# -------------------   Hold out    -------------------
# random_state is used to get same datasets for each run 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)  # train and test split

# -------------------  Decision tree   -------------------

print("\n####################################################\n")
print("1- DT Gain Ratio")

# HOLD OUT

#set the timer
timer_start = time.perf_counter()
# training the model
tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(X_train, Y_train)
# predict the classes
prediction = tree.predict(X_test)
#end the timer
timer_end = time.perf_counter() 
print("\nHold out used Decision Tree using Gain Ratio")
print(f"\n DT Gain done in {timer_end - timer_start:0.4f} seconds")
print("Accuracy: ", tree.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction))

#Confusion Matrix
disp = plot_confusion_matrix(tree, X_test, Y_test, cmap=plt.cm.Blues)
title="Dt1-Holdout"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# BAGGING
# define the model
timer_start = time.perf_counter()
model_bagging = BaggingClassifier(base_estimator=tree, n_estimators=10, bootstrap=True)
model_bagging.fit(X_train, Y_train)
prediction2 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBagging used Decision Tree using Gain Ratio")
print(f"\n DT Gain done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_bagging.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction2))

disp = plot_confusion_matrix(model_bagging, X_test, Y_test, cmap=plt.cm.Blues)
title="Dt1-Bagging"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# BOOSTING
timer_start = time.perf_counter()
model_boosting = AdaBoostClassifier(base_estimator=tree, n_estimators=10)
model_boosting.fit(X_train, Y_train)
prediction3 = model_boosting.predict(X_test)
timer_end = time.perf_counter()
print("\nBoosting used Decision Tree using Gain Ratio")
print(f"\n DT Gain done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_boosting.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction3))

disp = plot_confusion_matrix(model_boosting, X_test, Y_test, cmap=plt.cm.Blues)
title="Dt1-Boosting"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# CROSS VALIDATION
depth = []
# n_jobs: The number of CPUs to use to do the computation
# cv: Determines the cross-validation splitting strategy: 7-fold
timer_start = time.perf_counter()
prediction4 = cross_val_predict(tree, X_test, Y_test)
timer_end = time.perf_counter()
print("\nCross Validation used Decision Tree")
print(f"\n DT Gain done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", cross_val_score(estimator=tree, X=X, y=Y, cv=7, n_jobs=4).mean())
print("\n")
print(classification_report(Y_test, prediction4))

print("\n####################################################")
print("2- DT Gini")

# HOLD OUT
timer_start = time.perf_counter()
tree2 = DecisionTreeClassifier(criterion='gini')
tree2.fit(X_train, Y_train)
# predict the classes
prediction5 = tree2.predict(X_test)
timer_end = time.perf_counter() 
print("\nHold out used Decision Tree using Gini Index")
print(f"\n DT Gini done in {timer_end - timer_start:0.4f} seconds")
print("Accuracy: ", tree2.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction5))

disp = plot_confusion_matrix(tree2, X_test, Y_test, cmap=plt.cm.Blues)
title="Dt2-Holdout"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# BAGGING
timer_start = time.perf_counter()
model_bagging = BaggingClassifier(base_estimator=tree2, n_estimators=10, bootstrap=True)
model_bagging.fit(X_train, Y_train)
prediction6 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBagging used Decision Tree using Gini Index")
print(f"\n DT Gini done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_bagging.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction6))

disp = plot_confusion_matrix(model_bagging, X_test, Y_test, cmap=plt.cm.Blues)
title="Dt2-Bagging"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# BOOSTING
timer_start = time.perf_counter()
model_boosting = AdaBoostClassifier(base_estimator=tree2, n_estimators=10)
model_boosting.fit(X_train, Y_train)
prediction7 = model_boosting.predict(X_test)
timer_end = time.perf_counter()
print("\nBoosting used Decision Tree using Gini Index")
print(f"\n DT Gini done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_boosting.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction7))

disp = plot_confusion_matrix(model_boosting, X_test, Y_test, cmap=plt.cm.Blues)
title="Dt2-Boosting"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# -------------------     Naive Bayes  -------------------

print("\n####################################################\n")
print("3- Naive Bayes")

#HOLD OUT
timer_start = time.perf_counter()
naive_bayes = GaussianNB()
# train the model using the training sets
naive_bayes.fit(X_train, Y_train)
# predict the response for test dataset
predicted = naive_bayes.predict(X_test)
timer_end = time.perf_counter() 
print("\nHold out used Naive Bayes")
print(f"\n Naive Bayes done in {timer_end - timer_start:0.4f} seconds")
print("Accuracy:", metrics.accuracy_score(Y_test, predicted))
print("\n")
print(classification_report(Y_test, predicted))

disp = plot_confusion_matrix(naive_bayes, X_test, Y_test, cmap=plt.cm.Blues)
title="NaiveBayes-Holdout"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# BAGGING
timer_start = time.perf_counter()
model_bagging = BaggingClassifier(base_estimator=naive_bayes, n_estimators=10, bootstrap=True)
model_bagging.fit(X_train, Y_train)
prediction8 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBagging used Naive Bayes using Gaussion Distribution")
print(f"\n Naive Bayes done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_bagging.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction8))

disp = plot_confusion_matrix(model_bagging, X_test, Y_test, cmap=plt.cm.Blues)
title="NaiveBayes-Bagging"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# BOOSTING
timer_start = time.perf_counter()
model_boosting = AdaBoostClassifier(base_estimator=naive_bayes, n_estimators=10)
model_boosting.fit(X_train, Y_train)
prediction9 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBoosting used Naive Bayes using Gaussion Distribution")
print(f"\n Naive Bayes done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_boosting.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, prediction9))

disp = plot_confusion_matrix(model_boosting, X_test, Y_test, cmap=plt.cm.Blues)
title="NaiveBayes-Boosting"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# -------------------  Artificial Neural Network -------------------
print("\n####################################################\n")
print("4- Artificial Neural Network: 1 Hidden Layer")

#HOLD OUT
timer_start = time.perf_counter()
ann1 = MLPClassifier(hidden_layer_sizes=6, max_iter=1000)
ann1.fit(X_train, Y_train.values.ravel())
predictions = ann1.predict(X_test)
timer_end = time.perf_counter() 
print("\nHold out used Artificial Neural Network with 1 Hidden Layer:")
print(f"\n ANN 1 Layer done in {timer_end - timer_start:0.4f} seconds")
print("Accuracy", metrics.accuracy_score(Y_test, predictions))
print("\n")
print(classification_report(Y_test, predictions))

disp = plot_confusion_matrix(ann1, X_test, Y_test, cmap=plt.cm.Blues)
title="ANN1-Holdout"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# BAGGING
timer_start = time.perf_counter()
model_bagging = BaggingClassifier(base_estimator=ann1, n_estimators=10, bootstrap=True)
model_bagging.fit(X_train, Y_train)
predictions2 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBagging used ANN using 1 Hidden Layer")
print(f"\n ANN 1 Layer done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_bagging.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, predictions2))

disp = plot_confusion_matrix(model_bagging, X_test, Y_test, cmap=plt.cm.Blues)
title="ANN1-Bagging"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# BOOSTING
#It cannot be used any base estimator with AdaBoostClassifier
#Only those that support a "sample_weight" fit argument are allowed (e.g., DecisionTreeClassifier or SGDClassifier).


print("\n####################################################\n")
print("5- Artificial Neural Network: 2 Hidden Layer")

#HOLD OUT
timer_start = time.perf_counter()
ann2 = MLPClassifier(hidden_layer_sizes=(5,6,), max_iter=1000)
ann2.fit(X_train, Y_train.values.ravel())
predictions3 = ann2.predict(X_test)
timer_end = time.perf_counter() 
print("\nHold out used Artificial Neural Network with 2 Hidden Layer:")
print(f"\n ANN 2 Layer done in {timer_end - timer_start:0.4f} seconds")
print("Accuracy:", metrics.accuracy_score(Y_test, predictions3))
print("\n")
print(classification_report(Y_test, predictions3))

disp = plot_confusion_matrix(ann2, X_test, Y_test, cmap=plt.cm.Blues)
title="ANN2-Holdout"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# BAGGING
timer_start = time.perf_counter()
model_bagging = BaggingClassifier(base_estimator=ann2, n_estimators=10, bootstrap=True)
model_bagging.fit(X_train, Y_train)
predictions4 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBagging used ANN using 2 Hidden Layer")
print(f"\n ANN 2 Layer done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_bagging.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, predictions4))

disp = plot_confusion_matrix(model_bagging, X_test, Y_test, cmap=plt.cm.Blues)
title="ANN2-Bagging"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()

# BOOSTING
#It cannot be used any base estimator with AdaBoostClassifier
#Only those that support a "sample_weight" fit argument are allowed (e.g., DecisionTreeClassifier or SGDClassifier).


# -------------------   Support Vector Machine -------------------
print("\n####################################################\n")
print("6- Support Vector Machine")

#HOLD OUT

timer_start = time.perf_counter()
# Create a svm classifier
svm_model = svm.SVC(kernel='linear') # Linear Kernel
# Train the model using the training sets
svm_model.fit(X_train, Y_train)
# Predict the response for test dataset
Y_pred = svm_model.predict(X_test)
timer_end = time.perf_counter() 
print("\nHold out used SVM")
print(f"\n SVM done in {timer_end - timer_start:0.4f} seconds")
print("Accuracy:", metrics.accuracy_score(Y_test, Y_pred))
print("\n")
print(classification_report(Y_test, Y_pred))

disp = plot_confusion_matrix(svm_model, X_test, Y_test, cmap=plt.cm.Blues)
title="SVM-Holdout"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# BAGGING
timer_start = time.perf_counter()
model_bagging = BaggingClassifier(base_estimator=svm_model, n_estimators=10, bootstrap=True)
model_bagging.fit(X_train, Y_train)
pred2 = model_bagging.predict(X_test)
timer_end = time.perf_counter()
print("\nBagging used SVM")
print(f"\n SVM done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_bagging.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, pred2))

disp = plot_confusion_matrix(model_bagging, X_test, Y_test, cmap=plt.cm.Blues)
title="SVM-Bagging"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()


# BOOSTING
timer_start = time.perf_counter()
model_boosting = AdaBoostClassifier(base_estimator=svm_model, n_estimators=10, algorithm='SAMME')
model_boosting.fit(X_train, Y_train)
pred3 = model_boosting.predict(X_test)
timer_end = time.perf_counter()
print("\nBoosting used SVM")
print(f"\n SVM done in {timer_end - timer_start:0.4f} seconds")
print ("Accuracy:", model_boosting.score(X_test, Y_test))
print("\n")
print(classification_report(Y_test, pred3))

disp = plot_confusion_matrix(model_boosting, X_test, Y_test, cmap=plt.cm.Blues)
title="SVM-Boosting"
disp.ax_.set_title(title)
print(disp.confusion_matrix)
plt.show()
```
